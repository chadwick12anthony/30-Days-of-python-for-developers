```python
import requests
from requests_html import HTML
from selenium import webdriver 
from selenium.webdriver.safari.options import Options
from selenium.webdriver.common.by import By
import time

options = Options()
options.add_argument("--headless")
driver = webdriver.Safari(options=options)

url = "https://www.amazon.com/MERACH-Treadmill-Treadmills-0-5-7-5MPH-Brushless/dp/B0FR5C738N/ref=pd_sbs_d_sccl_1_7/139-4663855-0674236?pd_rd_w=rm7KD&content-id=amzn1.sym.aa738fbd-ad05-4d11-aae2-04b598db6305&pf_rd_p=aa738fbd-ad05-4d11-aae2-04b598db6305&pf_rd_r=1GKAC09T97ZV6JR535KV&pd_rd_wg=8EwwV&pd_rd_r=4f3811f8-0f49-4319-a0a4-6759df9ca654&pd_rd_i=B0DRY6WHW7&th=1"

driver.get(url)
time.sleep(20)

try:
    continue_button = driver.find_element(By.CSS_SELECTOR, "button.a-button-text[alt='Continue shopping']")
    continue_button.click()
    time.sleep(15)
except:
    pass

page_source = driver.page_source
html_obj = HTML(html=page_source)

title_lookup = '#productTitle'
price_lookup = '.a-price-whole'

title_element = html_obj.find(title_lookup, first=True)
price_element = html_obj.find(price_lookup, first=True)

product_title = title_element.text if title_element else "Title not found"
product_price = price_element.text if price_element else "Price not found"

print(product_title, product_price)

```
