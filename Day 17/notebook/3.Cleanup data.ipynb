```python
import os
import pandas as pd
```

```python
#Current working directory

cwd = os.getcwd() # dirname (abspath(file))
BASE_DIR = os.path.dirname(cwd)
DATA_DIR = os.path.join(BASE_DIR, 'data')
CACHE_DIR = os.path.join(BASE_DIR,'cache')
working_file = os.path.join(CACHE_DIR, 'movies-box-office-dataset.csv')
output_file = os.path.join(CACHE_DIR, 'movies-box-office-dataset-cleaned.csv')
```

```python
df = pd.read_csv(working_file)
df.head(n=1200)
```

```python
df["Domestic %"] = df['%']
df["Foreign %"] = df['%.1'] 
df.head()
```

```python
# Deleting columns 
# new_def = df.drop(columns=['%', '%.1'])   delete and set old df to the new one
df.drop(['%', '%.1'], axis=1, inplace=True) # delete completely from the old one
df.head()
```

```python

to_clean_cols = ['Worldwide', 'Domestic', 'Foreign']
                 
def currency_str_to_int(current_val):
    currency_val = current_val.replace("$", "").replace(",", "")
    try:
        currency_val = int(currency_val)
    except :
        # Takes any row value with "-" and turns into 0
        currency_val = 0
    return currency_val
        
def clean_col(row):
    """
    axis means we pass row by row 
    row = Pandas Series
    dataframe = collection of Pandas Series
    """
    #print(row)
    for col in to_clean_cols:
        current_val = row[col]
        row[col] = currency_str_to_int(current_val)
    return row
df_cleaned = df.apply(clean_col, axis=1)
df_cleaned.head()
```

```python
df.dtypes, df_cleaned.dtypes
```

```python
df_cleaned.sort_values(by=['Worldwide'], inplace=True, ascending=False)
df_cleaned.head()
```

```python
df_cleaned.reset_index(inplace=True, drop=True)
df_cleaned.head()
```

```python
# Sorting Rank using indices
df_cleaned['Rank'] = df_cleaned.index + 1
df_cleaned.head()
```

```python
df_cleaned['Domestic %'] = df_cleaned['Domestic'] / df_cleaned['Worldwide']
df_cleaned['Foreign %'] = df_cleaned['Foreign'] / df_cleaned['Worldwide']
df_cleaned.head(n=10)
```

```python
df_cleaned.to_csv(output_file, index=True)
```
