```python
def clean_data_ (text, keyname = None):
    if keyname == "vote":
        return text.replace(" votes", "")
    if keyname == "tag":
        return text.replace("\n", "/")
    else :
        return text
```

```python
def parsed_tagged_pages(html_text):
    question_summaries = html_text.find(".js-post-summary")
    needed_classes = ['.s-link','.s-post-summary--stats-item', '.s-post-summary--meta-tags']
    key_names = ["question","vote", "tag"]
    data = []
    for q_el in question_summaries:
        data_names = {}
        for i , _class in enumerate(needed_classes):
            class_el = q_el.find(_class)[0]
            key_name = key_names[i]
            data_names[key_name] = clean_data_(class_el.text, keyname=key_name)
        data.append(data_names)
    return data
    
```

```python
def get_data_from_url_(url):
    r =requests.get(url)
    if r.status_code not in range (200,299):
        return []
    html_str = r.text
    html_text = HTML(html=html_str)
    data = parsed_tagged_pages(html_text)
    return data
```

```python
base_url = 'https://stackoverflow.com/questions/tagged/'
tag = "javascript"
query_filter = "Votes"
url = f"{base_url}{tag}?tab={query_filter}"
get_data_from_url_(url)
# print(url)
```

```python
def get_data_from_pages_(tag = "javascript", query_filter = "votes", max_p =5, pagesize=25):
    data = []
    base_url = 'https://stackoverflow.com/questions/tagged/'
    for p in range (max_p):
        page_nbr = p+1
        #url = https://stackoverflow.com/questions/tagged/python?tab=votes&page=2&pagesize=15
        url = f"{base_url}{tag}?tab={query_filter}&page={page_nbr}&pagesize={pagesize}"
        data += get_data_from_url_(url)
        time.sleep(1.2)
    return data
#get_data_from_pages_()
```
